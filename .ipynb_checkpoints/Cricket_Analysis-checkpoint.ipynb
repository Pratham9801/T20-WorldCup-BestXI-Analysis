```python
df_batting['match_key']=df_batting['match'].str.split('Vs').apply(lambda x : ' vs '.join(sorted([i.strip() for i in x])))

#df_batting['match'].str.split(' Vs ')
#This splits the string "Namibia Vs Sri Lanka" into a list of two teams:

#sorted(x)
#Sorts the list alphabetically

# ' vs '.join(...)
#Joins the sorted list back into a single string

#The .join() method in Python inserts the given string between the elements of a list and returns a single string.

#Think of ' vs ' as the glue you're using to combine the list items.
```

```python
df_match['match_key']=df_match.apply(lambda row: ' vs '.join(sorted([row['team1'],row['team2']])),axis=1)
```

```python
#Let us create a dict and with help of that dict let us join both the dstaframes 

match_ids_dict ={}
for index,rows in df_match.iterrows():   #iterrows(): used to iterate each row 
    print(index, rows)
```

```python
for index,rows in df_match.iterrows():
    
    team1 = rows['team1'].strip()
    team2 = rows['team2'].strip()
    
    key1 = f"{team1} Vs {team2}"
    key2 = f"{team2} Vs {team1}"
    
    
    match_ids_dict[key1]=rows['match_id']
    match_ids_dict[key2]=rows['match_id']
    
match_ids_dict   
```

```python
match_ids_dict['Namibia Vs Sri Lanka']    # give sus the matchid for that match 
```

```python
df_batting['mach_id']=df_batting['match'].map(match_ids_dict)

df_batting.head(10)

# The .map() function is used to transform or map values 
# in a pandas Series (i.e., a column) based on a dictionary, function, or another Series.

#Look at each value in the 'match' column of df_batting

#Use that value as a key to look it up in the dictionary match_ids_dict

#Return the value from the dictionary (which is the match_id)

#Create a new Series with those match_id values
```

### Bowler’s Economy Rate (Economy)

Economy rate measures how many runs a bowler concedes per over.

Economy Rate= Total Runs Conceded / Overs Bowled 
​
 
🏏 Example:
Bowler gives 40 runs in 8 overs.

Economy = 40 ÷ 8 = 5.00

Lower Economy → Better control (bowler is giving fewer runs per over)

Especially important in limited overs cricket (T20, ODI) where run containment is crucial.

Sometimes even a bowler with fewer wickets but a great economy is more valuable.
